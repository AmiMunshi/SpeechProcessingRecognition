```python
#Know the parameters of  of the audio
audio_path="/content/hello.wav"

# using x variable of audio amplitude in numbers, sr is the sampling rate
x,sr= librosa.load(audio_path)
print(x)
print(sr)
```

```python
#Import matplotlib and librosa display to see the plot
import matplotlib.pyplot as plt
import librosa.display
```

```python
plt.figure(figsize=(15,5))
librosa.display.waveshow(x,sr=sr)
```

```python
X= librosa.stft(x) #convert audio signal to short term fourier transform
XdB= librosa.amplitude_to_db(X)
plt.figure(figsize=(15,5))
librosa.display.specshow(XdB, sr=sr, x_axis= "time", y_axis="hz")
plt.colorbar()
```
